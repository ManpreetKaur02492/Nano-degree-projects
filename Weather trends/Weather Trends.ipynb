As per above Visualization, There is no missing values in Global_data but there are 3 missing values in column of avg_temp in Victoria_data.This is not a big deal beacuse it doesn't affect the results.So, there is no need of data cleaning.

```python
Victoria_data2=Victoria_data.dropna()
Victoria_data2.head(15)
```

```python
Victoria_data2=Victoria_data.fillna(Victoria_data.mean())  # we can fill NaN values with Mean value(mean temp)
Victoria_data2
```

```python
fig = plt.figure(figsize = (20,10))
plt.suptitle('Global vs. Victoria City Temperature Trend', fontsize =25,Color='green' )
plt.plot(Global_data.year, Global_data.avg_temp,color='red',label='Global_temp')
plt.axhline(Global_data.avg_temp.mean(), color='b', linestyle='dotted', linewidth=3,label='Global_avg_temp_mean')
plt.legend(loc=0)
plt.xlabel('Year')
plt.ylabel('Temperature(℃)')
plt.xticks(np.arange(1750,2020,step=5), rotation=45)

plt.plot(Victoria_data2.year, Victoria_data2.avg_temp,color='orange', label='Victoria_temp')
plt.axhline(Victoria_data2.avg_temp.mean(), color='k', linestyle='dotted', linewidth=3,label='Victoria_avg_temp_mean')
plt.legend(loc=0)
plt.show()

```

The above visualiztion showed the great insights to data.Global'temp data starts from year 1750 but city Victoria's data starts from year 1828.There are peaks into both data but Victoira's temp data has more sharp peaks which means It changed sharply along with time as compared to Global temp.

# Moving Averages

The idea behind the moving averages is to smoother out the values and reduce the noise. It makes easier to watch out the trends more clearly with moving averages. I calculated moving averages for 5, 10 and 15 years because it makes visualization clearer to analyze.

```python
n = 5

Global_data['moving_avg5'] = Global_data['avg_temp'].rolling(window=n).mean()
Victoria_data2['moving_avg5'] = Victoria_data2['avg_temp'].rolling(window=n).mean()

n = 10

Global_data['moving_avg10'] = Global_data['avg_temp'].rolling(window=n).mean()
Victoria_data2['moving_avg10'] = Victoria_data2['avg_temp'].rolling(window=n).mean()

n = 15

Global_data['moving_avg15'] = Global_data['avg_temp'].rolling(window=n).mean()
Victoria_data2['moving_avg15'] = Victoria_data2['avg_temp'].rolling(window=n).mean()

plt.figure(figsize=(20,10))
plt.plot(Global_data.year, Global_data.avg_temp, label='yearly_avg')
plt.plot(Global_data.year, Global_data.moving_avg5, label='moving_avg5')
plt.plot(Global_data.year, Global_data.moving_avg10, label='moving_avg10')
plt.plot(Global_data.year, Global_data.moving_avg15, label='moving_avg15')
plt.title('Global Weather Trends')
plt.xlabel('Year')
plt.xticks(np.arange(1750, 2020, step=5), rotation=45)
plt.ylabel('Temperature(℃)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(Victoria_data2.year, Victoria_data2.avg_temp, label='yearly_avg')
plt.plot(Victoria_data2.year, Victoria_data2.moving_avg5, label='moving_avg5')
plt.plot(Victoria_data2.year, Victoria_data2.moving_avg10, label='moving_avg10')
plt.plot(Victoria_data2.year, Victoria_data2.moving_avg15, label='moving_avg15')
plt.legend()
plt.title('Victoria Weather Trends')
plt.xlabel('Year')
plt.xticks(np.arange(1820, 2015, step=5), rotation=45)
plt.ylabel('Temperature(℃)')
plt.show()
```
